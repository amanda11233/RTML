## Methods
### AlexNet
For the AlexNet model, we have tried two methods (Sequential API and Module API)
The Sequential API is a sraightforward approach for the model which is just a sequence of layers inside a pytorch sequential layer
It is a relatively simple way of creating a deep learning model.

The Module API is more flexible and modular approach for the model. It allows multiple inputs and ouputs. 
The training of the AlexNet model was done for 10 epochs with learning rate of 0.001 and momentum of 0.9, the loss function used is CrossEntropy Loss Function 
and the optimizer used is SGD (Stochasti Gradient Decent)

For the googlenet, we have experimented with implementation of the googlenet model from the scratch and also the pytorch pretrained library.
The training of the googlenet is done for  25 epoch for scratch and 10 epoch for pretrained. The learning rate is 0.001, momentum is 0.9, loss function is CrossEntropy loss function and the optimizer is Adam.

The dataset experimented on is the CIFAR-10 dataset. 

For both alexnet and googlenet inception model implemenation from the scratch, we have followed the same architecture mentioned in the respective papers.
